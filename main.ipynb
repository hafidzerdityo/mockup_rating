```python
from faker import Faker
fake = Faker()
```

```python
import pandas as pd
from datetime import datetime

gas = 364
tgl  = pd.date_range(
    start=pd.Timestamp('2022-12-31') - pd.offsets.Day(gas),
    freq='D', periods=gas)
tgl 
```

```python
kategori = {
    1: ['paket rusak', 'salah alamat', 'lama'],
    2: ['kurir jutek', 'tidak komunikatif'],
    3: ['nothing special', 'kecepatan pengiriman biasa aja'],
    4: ['lumayan cepat', 'lumayan ramah', 'waktu normal'],
    5: ['pengiriman cepat', 'kurir ramah', 'tepat waktu', 'barang aman']
}
```

```python
import pymongo
import pandas as pd

connection_string = "mongodb://localhost:27017"  # local

# mongo db cloud
# connection_string = f'mongodb+srv://hafidz_erdityo:lelegoreng@cluster0.uorwk7c.mongodb.net/?retryWrites=true&w=majority'

client = pymongo.MongoClient(connection_string)
```

```python
get_data_wilayah = list(client['wilayah']['provinsi'].find({}))
get_data_wilayah = pd.DataFrame(get_data_wilayah)
```

```python
get_data_wilayah
```

```python
big_str = ''

for i in get_data_wilayah['Provinsi'].to_list():
    big_str += i + '\n'
```

```python
with open('data_provinsi.txt', 'w') as f:
    f.write(big_str)
```

```python
with open('data_provinsi.txt') as f:
    data_provinsi = f.read()
data_provinsi = data_provinsi.split('\n')
data_provinsi[:3]
```

```python
daftar_nama = []

for _ in range(50):
    daftar_nama.append(fake.name())
daftar_nama[:3]
```

```python
import random
```

```python
kantor = [
    'Kantor A',
    'Kantor B',
    'Kantor C',
    'Kantor D',
    'Kantor E'
]
```

```python
data = []


cnt = 0
for _ in range(100):
    random_tgl = random.choice(tgl)
    each_data = {}
    each_data['id'] = cnt
    each_data['nama'] = random.choice(daftar_nama)
    each_data['tanggal'] = random_tgl
    each_data['rating'] = random.choice(range(1,6))
    each_data['kategori'] = random.choice(kategori.get(each_data['rating']))
    each_data['wilayah'] = random.choice(data_provinsi)
    each_data['kantor'] = random.choice(kantor)
    data.append(each_data)
    cnt += 1
```

```python
import pandas as pd

x = pd.DataFrame(data)
x
```

```python
x.to_csv('dummy.csv')
```
